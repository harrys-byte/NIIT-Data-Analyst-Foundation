```python
# TASK1

def palindrome():
    a=input("Enter String : ")
    
    if a.upper()==a[::-1].upper():
        print("PALINDROME")
    else:
        print("NOT A PALINDROME")
    
palindrome()

'''
A user-defined function palindrome() has been created.
It accepts a string from the user and checks if the original string
is equal to its reversed form.
If True it returns 'PALINDROME' else 'NOT A PALINDROME'
'''
```

```python
# TASK2

def sortd(limit):
    L=[]
    for i in range(limit):
        s=input("Enter Word : ")
        L.append(s)
    
    L.sort()
    return L

limit=int(input("Enter Number of Words : "))
sortd(limit)

'''
The number of words is accepted from the user in the form of integer.
A user-defined function sortd() has been created.
An empty list L has been created.
Then using for loop and range() we are accepting a number of words of the
user's choice a nd the elements are added to the list L using
append().
Then the list is sorted using sort() and is displayed using return keyword.
'''
```

```python
# TASK3

def convert():
    L=input("Enter List : ").split(',')
    
    b=''.join(L)
    
    print('String Format :',b)    

convert()

'''
A user-defined function convert() has been created.

It accepts a list in the form of string and is then converted to list
using split().

Then we have converted the list to string using join() and the output
is displayed using print().
'''
```

```python
# TASK4

def inlist():
    print("MAXIMUM INPUT LIMIT : 10\n")
    L=[]
    
    for i in range(10):
        a=int(input("Enter Number : "))
        L.append(a)
    
    print("Final List :",L)
    print("\nLargest Number :",max(L))
    print("Smallest Number :",min(L))
    
inlist()
```

```python
# TASK5

def maxmin():
    print("MAXIMUM INPUT LIMIT : 5\n")
    d={}
    
    for i in range(5):
        a=float(input("Enter Value : ")
        
        d['Item'+str(i+1)]=a
        
    print("\nFinal Dictionary :\n")
    for i,j in d.items():
        print(i,':',j)
        
    print("\nMaximum Value :",max(d.values()))
    print("Minimum Value :",min(d.values()))

maxmin()
```

```python
# TASK6

def f1():
    s=[]
    while True:
        n=input("\nEnter Name : ")
        a=int(input("Enter Age : "))
        g=input("Enter Gender : ")
        o=input("Enter Occupation : ")
        
        d={}
        d['Name']=n
        d['Age']=a
        d['Gender']=g
        d['Occupation']=o
        s.append(d)
        
        ans=input("\nAdd another record? YES/NO : ")
        if ans.upper()=='NO':
            break;
            
    return s

uprec=f1()

def f2():
    print('\nFINAL RECORDS :')
    for i in uprec:
        print('\n')
        for j,k in i.items():
            print(j,':',k)
f2()
```
