```python
# TASK3

numlist=list((range(1,30)))
print(numlist)

counteven=countodd=0

for i in numlist:
    if i%2==0:
        counteven+=1
    else:
        countodd+=1

print("Number of Even Numbers :",counteven)
print("Number of Odd Numbers :",countodd)

'''
We have a list named numlist assigned the sequence of values from 1 to 29.
We have assigned the value 0 to two variables counteven and countodd.

Using for loop we have sorted the even and the odd numbers and have counted the number
of even and odd numbers seperately using if else statements.

Then we have printed the count of even and odd numbers individually using print().

count() : Returns the count of the element in the list.
Membership Operator IN : Used to check the existence of a element in a sequence of datatypes.
'''

# TASK4

NewL=[]
for i in numlist:
    if i%2==0:
        if i<11:
            NewL.append(i)
            
print("First 5 Even Numbers :",NewL)

'''
We have have created an empty list NewL.

Using for loop we have sorted out the even numbers and have appended the first 5 even numbers
into the NewL list using append().

Then we have printed the NewL using print().

append() : Adds the given element to the given list.
for loop : Used when the extreme and step values are known.
Membership Operator IN : Used to check the existence of a element in a sequence of datatypes.
'''
        
# TASK 5

for i in numlist:
    if i%2==0:
        numlist.remove(i)
        
print("Last Two Odd Number :",numlist[-1],numlist[-2])

'''
Using for loop we have sorted out the even numbers from the 'numlist' list and removed them
from the list in order to retrieve a list containing only odd numbers.

This is done as we are restricted from creating a new list or reversing the list.

Then using the index of the last two elements we have printed them using print().

for loop : Used when the extreme and step values are known.
remove() : Removes the given element from the given list.
Membership Operator IN : Used to check the existence of a element in a sequence of datatypes.
'''
```

```python
# TASK6

x=[]
y=[]

for i in range(5):
    a=input("Enter String to insert in LIST x : ")
    x.append(a)

print('\nx =',x,'\n')

for i in range(3):
    b=int(input("Enter Number to insert in LIST y : "))
    y.append(b)
    
print('\ny =',y,'\n')

for i in y:
    x.append(i)

print("FINAL LIST :",x)

'''
The above query returns the final list as the concatenation of the
two user-defined lists.

Two empty lists named x and y are created.
We accept 5 strings and 3 numbers from the user and store them in
lists x and y respectively using the append().

Then the elements of the list y are appended to list x using the append() in for loop.

append() : Adds the given element to the given list.
range() : (start,stop,step) Used in for loop to iterate through the given range of values.
for loop : Used when the extreme and step values are known.
Membership Operator IN : Used to check the existence of a element in a sequence of datatypes.
'''
```
