```python
import numpy as n
import pandas as p
import matplotlib.pyplot as mpl

df=p.read_excel('DS1_C4_S5_Employee_Data_Practice.xlsx')
df
```

```python
df['Department'].unique()
```

```python
df['Department'].nunique()
```

```python
gr=df.groupby(by='Department')['Annual Salary ($)'].sum()
gr
```

```python
mpl.figure(figsize=(12,6))
mpl.title('SUM OF SALARIES : DEPARTMENT-WISE')
mpl.bar(gr.index,gr.values,color='Maroon')
mpl.xlabel('DEPARTMENT')
mpl.ylabel('SALARY')
mpl.show()

'''
The modules numpy, matplotlib.pyplot and pandas have been imported.
Using groupby() we have grouped the data of the employees who belonging to 4 different
departments.
Then we have used the Annual Salary column to find the sum of salaries of all the employees.
Using this, we have used visualized the grouped data in the form of Bar Graph using bar().
'''
```

```python
# TASK3

av=df[(df['Work_Experience']==0) & (df['Department']=='IT ')]['Annual Salary ($)'].mean()

print("Estimated Salary for a Fresher :",av)

'''
We have filtered out the data of the employees whose work experience is 0 and belong
to the IT department.
Then we have used the Annual Salary column to find the average of the filtered employees using
mean() and has been printed using print().
'''
```

```python
# TASK4

gr=df.groupby(by='Department')['Annual Salary ($)'].sum()
print(gr)
gr.plot(kind='pie',autopct='%.2f%%')
mpl.show()

'''
We have used the groupby() and have grouped the data of all employees into their
respective departments and have found the sum of salaries of the employees using the
Annual Salary column.
We have used the plot() so as to display the grouped data in the form of Pie Chart.'''
```
