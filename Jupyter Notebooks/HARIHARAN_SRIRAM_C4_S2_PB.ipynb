```python
# TASK2

import pandas as py
import numpy as n

rollno=n.arange(1001,1041)

phy=n.random.randint(30,100,40)
mat=n.random.randint(30,100,40)

pphy=py.Series(data=phy,index=rollno)
pmat=py.Series(data=mat,index=rollno)

print('PHYSICS MARKS :')
print(pphy)

print('\n')
print('MATHEMATICS MARKS :')
print(pmat)

'''
The pandas and the numpy libraries have been imported.

Random marks have been assigned using random module and randint().
Two pandas series variable have been created for both Physics and Mathematics seperately
and have been printed using print().
'''
```

```python
# TASK3

print('SUM OF PHYSICS AND MATHEMATICS MARKS:')
sumpm=pphy+pmat
print(sumpm)

'''
The sum of Physics and Mathemetics marks have been printed using print().
'''
```

```python
# TASK 4

df=py.DataFrame()
df['STUDENT ID']=rollno
df['MATHEMATICS MARKS']=pphy
df['PHYSICS MARKS']=pmat
df['SUM']=phy+mat

qs=df['STUDENT ID'][df['SUM']>120]
print('TOTAL NUMBER OF STUDENTS WHO HAVE QUALIFIED :')
print(qs.size)
print('\nIDs OF QUALIFIED STUDENTS :')
print(qs)

'''
An empty dataframe has been created to tabulate the given data.
The IDs of qualified students i.e., the students whose sum of Physics and Mathematics marks
are greater than 120 have been displayed using size and print().
'''
```

```python
# TASK5

df=py.DataFrame()
df['STUDENT ID']=rollno
df['ROUND1 QUALIFIERS']=phy+mat

qs=df[['STUDENT ID','ROUND1 QUALIFIERS']][df['ROUND1 QUALIFIERS']>120]

print(qs)

```
