```python
import numpy as n
import seaborn as sb
import matplotlib.pyplot as plt
import pandas as p


sach=p.read_csv('Sachin ODI Data Analysis Cleaned.csv')

sach=sach.rename(columns={'FullYear':'Year'})

'''
The libraries NUMPY,SEABORN,MATPLOTLIB and PANDAS have been imported for various mathematical,
statistical and visualization purposes.

The cleaned csv file has been read using read_csv().
In the data, the column 'FullYear' has been renamed to 'Year' using rename().
'''
sach
```

## TASK1
### Display the Frequency of 6s hit by Sachin Tendulkar

```python
# TASK1

print(sach.columns)

plt.figure(figsize=(12,6))
plt.title('FREQUENCY OF 6s HIT BY SACHIN TENDULKAR')
s4=sb.countplot(data=sach,x='6s',palette='Blues_r')

for i in s4.containers:
    s4.bar_label(i)

plt.show()

'''
The columns of the data has been printed.

The length and breadth of the graph has been modified using figure().
We have used the countplot() from the Seaborn library in order to display the data in the form
of BarGraph.

The datalabels of the graph has been made visible by using for loop,containers and bar_label().

The unnecessary data has been neglected using show().

Thus we can conclude that Sachin has not hit any sixes in 231 matches and 7 sixes in only
1 match.
'''
```

## TASK3
### Display year-wise number of matches played by Sachin Tendulkar

```python
# TASK3

plt.figure(figsize=(10,8))
plt.title('YEAR-WISE NUMBER OF MATCHED PLAYED BY SACHIN')
s=sb.countplot(data=sach,y='Year')

for i in s.containers:
    s.bar_label(i)

plt.show()

'''
The length and breadth of the graph has been modified using figure().
We have used the countplot() from the Seaborn library in order to display the data in the form
of BarGraph.
The datalabels of the graph has been made visible by using for loop,containers and bar_label().
The unnecessary data has been neglected using show().

Thus we can conclude that, Sachin has played the most number of matches in the year 2005 (25)
and the least number of matches in the year 1989 (1).
'''
```

## TASK4
### Display the number of 50s hit by Sachin Tendulkar against every opponent team

```python
# TASK4

sac=sach[sach['50']==1]

p.crosstab(index=sac['Opposition'],
           columns=sac['50']).sort_values(by=1,ascending=True).plot(kind='barh',figsize=(12,6),
                                                            title='NUMBER OF 50s HIT BY SACHIN',
                                                                   xlabel='Number of 50s')
plt.show()

'''
The given data has been filtered into the data containing only the matches in which Sachin
has scored one or more than one 50s.

Then we have used crosstab() retrieve the dataframe where we have the grouped data
according to the each opposition teams.
Then we have sorted out the data using sort_values() and plot() to visualize the data
using a Horizontal bar graph.

Thus we can conclude that the Sachin has hit the most sixes against Australia (16) and least
against Bangladesh (1)
'''
```

## TASK8

### Display a correlation matrix between 4s, 6s and Position.

```python
# TASK8

cr=sach[['4s','6s','Pos']].corr()
plt.figure(figsize=(8,6))
plt.title('CORRELATION MATRIX BETWEEN 4s, 6s AND POSITION')
print(cr)

sb.heatmap(cr,annot=True,cmap='Oranges_r',cbar=True)

'''
We have considered the required columns '4s', '6s' and 'Pos' from the given data.
Then we have used a correlation matrix using corr() and have displayed it using print().

Then we have used heatmap() in order to display the data in the form of a Heat Map with a
intensity bar of correlation alongside of it.

Thus we can conclude that there is a,
MODERATE POSITIVE correlation between 6s and 4s.
NO CORRELATION between Position and 6s.
WEAK NEGATIVE correlation between Position and 4s.
'''
```
