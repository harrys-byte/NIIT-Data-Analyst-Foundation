```python
a,b,c=4,3,2 # The values 4,3 and 2 are assigned to the variables a,b and c respectively

# TASK10 1
d=a**b+c # The **(exponential operator) has more priority than + (sum operator)
         # Thus it calculates a**b first and then adds the result with c
    
print(d) # The value of d is printed

# TASK10 2
e=a**(b+c) # Even though the **(exponential operator) has more priority than + (sum operator),
           # using parentheses will change the preference for the operators and
           # the order of preference for operators will not be followed
        
print(e) # The value of e is printed

# TASK10 3
'''
In the first task, the **(exponential operator) has more priority than + (sum operator).
Thus it calculates a**b first and then adds the result with c.

Whereas in the second task, even though the **(exponential operator) has more priority than 
+ (sum operator), using parentheses will change the preference for the operators as parentheses
has more priority than any other operator.
'''
```
