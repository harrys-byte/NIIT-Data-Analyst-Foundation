```python
import numpy as n
import pandas as p
import seaborn as sb
import matplotlib.pyplot as plt

maind=p.read_csv('IPL.csv')

maind

'''
The libraries NUMPY,SEABORN,MATPLOTLIB and PANDAS have been imported for various mathematical,
statistical and visualization purposes.

The given csv file has been read using read_csv() and the data has been displayed.
'''
```

```python
maind['date']=p.to_datetime(maind['date'])
maind['Year']=maind['date'].dt.year

maind

'''
In order to execute some questions we have to use the 'date' column.
But the required column is not in the proper date format.
So the 'date' column has been converted to date format using to.datetime().
From that we have extracted the year and created another column named 'Year' containing only
the year as its values.
'''
```

## Q1 
### How many matches were played from the year 2008 to 2020?


```python
s=maind[maind['Year']>=2008].sort_values(by='Year',ascending=False)

plt.figure(figsize=(12,6))
plt.title('NUMBER OF MATCHED PLAYED BETWEEN THE YEARS 2008 AND 2020')

a=sb.countplot(data=s,y='Year')

for i in a.containers:
    a.bar_label(i)
    
plt.show()

c=s.groupby(by='Year')['Year'].count()[0:].sum()

print('Total Number of Matches played between the years 2008 and 2020 :',c)

'''
The length and breadth of the graph has been modified using figure().
The title of the graph has been modified using the title().

We have used the countplot() from the Seaborn library in order to display the data in the form
of BarGraph.

The datalabels of the graph has been made visible by using for loop,containers and bar_label().

The unnecessary data has been neglected using show().
'''
```

#### INTERPRETATION :

##### 1. Total Number of matches played between the years 2008 and 2020 is 816.
##### 2. The most number of matches have been played in the year 2013 (76).
##### 3. The least number of matches have been played in the year 2009 (57).

## Q3
### Which city hosted the greatest number of matches?

```python
plt.figure(figsize=(10,10))
plt.title('NUMBER OF MATCHES PLAYED IN EACH CITY ')

d=sb.countplot(data=maind,y='city',color='yellow')

for i in d.containers:
    d.bar_label(i)
    
plt.show()

'''
The length and breadth of the graph has been modified using figure().
The title of the graph has been modified using the title().

We have used the countplot() from the Seaborn library in order to display the data in the form
of BarGraph.

The datalabels of the graph has been made visible by using for loop,containers and bar_label().

The unnecessary data has been neglected using show().
'''
```

#### INTERPRETATION :

##### 1. The city which hosted the greatest number of matches is MUMBAI (101).
##### 2. The city which hosted the least number of matches is Bloemfontein (2).

## Q4

### Present the top 5 players who won the most 'Man Of The Match' awards

```python
mm=maind['player_of_match'].value_counts()[:5].sort_values(ascending=True)
print(mm)

m=mm.plot(kind='bar',figsize=(12,6),
        title='TOP 5 PLAYERS WITH THE MOST --MAN OF THE MATCH-- AWARDS',
        xlabel='Number of Awards',ylabel='Players',color='maroon')

for i in m.containers:
    m.bar_label(i)
    
plt.show()

'''
The given data has been filtered out using the column 'Player of the Match'.
Value_counts() and sort_values() have been used to display the top 5 players from the
column 'Player of the Match'.

Using plot() the length, breadth and title of the graph have been modified.

We have used the countplot() from the Seaborn library in order to display the data in the form
of BarGraph.

The datalabels of the graph has been made visible by using for loop,containers and bar_label().

The unnecessary data has been neglected using show().
'''
```

#### INTERPRETATION :

##### The the top 5 players who won the most 'Man Of The Match' awards are
##### 1. AB de Villiers (23) 
##### 2. CH Gayle (22)
##### 3. RG Sharma (18)
##### 4. MS Dhoni (17)
##### 5. DA Warner (17)

## Q5

### Which team won the match with the highest result margin? Print the match details.

```python
rm2=maind['result_margin'].max()
print('Highest Result Margin :',rm2)

det=maind[maind['result_margin']==rm2]
print('Name of the Team :',list(det['winner'])[0])
```

#### INTERPRETATION :

##### 1. The highest result margin is 146.0.
##### 2. Team has won with the highest result margin is Mumbai Indians against Delhi Daredevils.
