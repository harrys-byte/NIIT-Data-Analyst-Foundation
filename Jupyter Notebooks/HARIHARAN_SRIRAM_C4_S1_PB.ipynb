```python
# TASK1

import numpy as n

al=n.arange(1,31)

hr=al[:3]
it=al[4:18]
fi=al[19:23]
sa=al[24:30]

print('Employee IDs :')
print('TOTAL :',al)
print('\nHR :',hr)
print('IT :',it)
print('FINANCE :',fi)
print('SALES :',sa)

'''
NUMPY has been imported.
A variable 'al' has been assignned the values ranging from 1 to 30 in the form or 1D array
using arange() and array().
Using slicing those values have been segregated or grouped into 4 different values and have been
printed.

array() : Stores the value in the form 1D array.
arange() : Stores the given range of values in the form 1D array where
           Start value is inclusive and End value is exclusive.
'''
print('\n'*2)

# TASK2
hrsal=n.array([50,60,57])
itsal=n.full(15,67)
fisal=n.array([60,65,50,70,87])
sasal=n.array([50,60,65,90,56,65,71])

print('Salary (in thousands) :')
print('HR :',hrsal)
print('IT :',itsal)
print('FINANCE :',fisal)
print('SALES :',sasal)

'''
Given values have been assigned to 4 different variables in the form 1D array using
array() and full() and have been printed.

array() : Stores the value in the form 1D array.
full() : Assigns the same given value for n number of  elements. 
'''
print('\n'*2)

# TASK3

hrav=n.mean(hrsal)
itav=n.mean(itsal)
fiav=n.mean(fisal)
saav=n.mean(sasal)

print('Average Salary (in thousands) :')
print("HR :",n.round(hrav,2))
print("IT :",itav)
print("FINANCE :",fiav)
print("SALES :",n.round(saav,2))

L=n.array([hrav,itav,fiav,saav])
print('\nDepartment with Highest Average Salary (in thousands) is IT with',n.max(L))

'''
The mean of the values created in task2 has been found for each department seperately using
mean().
These values have stored in a variable 'L' in the form of 1D array using array().
Then the maximum of the four values has been found out using max().

array() : Stores the value in the form 1D array.
mean() : Can be used after imported NUMPY to find the average of the given values.
'''
print('\n'*2)

# TASK4

print("Highest Salary (in thousands) :")
print("HR :",max(hrsal))
print("FINANCE :",max(fisal))

'''
We have found out the maximum salary of the HR and FINANCE departments individually
using max().
'''
print('\n'*2)

# TASK5

print('Updated Salary (in thousands) for IT employees :\n',itsal*1.05)

'''
We have updated the original salaries of the employees working in the IT department by
incrementing 5% of their individual gross salary to their original salaries.
'''
```
