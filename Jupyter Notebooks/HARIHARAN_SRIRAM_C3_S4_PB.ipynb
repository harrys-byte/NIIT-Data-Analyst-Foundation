```python
# TASK5

t1=(1,2,3)
t2=(4,5,6)

print('BEFORE :')
print('ID of t1 :',id(t1))
print('ID of t2 :',id(t2))

t1=t1+t2

print('\nAFTER :')
print('ID of t1 :',id(t1))
print('ID of t2 :',id(t2))


'''
We have created two tuples namely t1 and t2 and have assigned some values
to it.
Then we have printed the memory address of both of the tuples using
id() and print().

Then we have concatenated the tuples t1 and t2 and assigned it to a 
variable t1.
Then again we have printed the memory address of both of the tuples using
id() and print().

From the output we can conclude that the memory address of the tuple t2
remains the same whereas the memorya address of the tuple t1 has changed
because its value has been updated in the expression t1=t1+t2.
'''
```

```python
# TASK6

m={10,20,30,40,50}
n=[33,40,55,60,70]

for i in range(len(n)):
    m.add(n[i])
    
print('FINAL SET m :',m)

'''
A set 'm' and a list 'n' has been created and has been assigned 
some values.

Then we have used the for loop, range(), len() and add() to add the elements
of n one-by-one into the set m.

Then we have printed the set 'm' using print().

The set will ignore duplicate values because sets can conatain only
unique or distinct elements.

for loop : Used when the extreme and step values are known.
Membership Operator IN : Used to check the existence of a element in a sequence of datatypes.
add() : Adds the passed value into a set.
range() : (start,stop,step) Used in for loop to iterate through the given range of values.
len() : Returns the length of the sequence in th form of integer.
'''
```
