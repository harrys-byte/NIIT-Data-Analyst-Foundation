```python
# TASK1 

a = input("Enter Character : ") # Accepts a string character and stores it in variable 'a'

if a=='A': # Checks if the value of a is equal to the value 'A'
    print("Python") # Prints if the if condition is True
elif a=='B': # Checks if the value of a is equal to the value 'B'
    print("MySQL") # Prints if the elif condition is True
elif a=='C': # Checks if the value of a is equal to the value 'C'
    print("Excel") # Prints if the elif condition is True
else: # Executes if the if condition is False
    print("Holiday") # Prints if the else condition is True
    
'''
We are accepting a character from the user and if it is 'A','B' or 'C' the output will be
'Python','MySQL' or 'Excel'. If the input is other than these, then the output will be
'Holiday'
'''
```

```python
# TASK2

a = int(input("Enter Number 1 : ")) # Accepts a number from the user and stores it in variable 'a' 
b = int(input("Enter Number 2 : ")) # Accepts a number from the user and stores it in variable 'b'

if a>b: # if : Checks if the value of a is greater than the value of b
    print("\nThe Greatest Number is",a) # Prints if the if condition is satisfied
elif b>a: # elif : Executes if the if condition is false and checks if b is greater than a
    print("\nThe Greatest Number is",b) # Prints if the elif condition is satisfied
else: # Executes if the elif condition returns False
    print("Both the numbers are Equal") # Prints if the else condition is satisfied
    
'''
We are accepting two numbers from the user.
Then we are printing the greatest of the two numbers or equal using comparison operators, 
if else and elif conditional statements.
'''
    
```

```python
# TASK3

mbs = float(input("Enter Monthly Basic Salary ($) : ")) # Accepts a float value from the user

if mbs>=0 and mbs<4000: # if : Checks if the value of mbs satisfies the conditions 
                        # greater or equal to $0 and lesser than $4000 using AND operator
        
    hra=(10/100)*mbs # 10% of mbs is calculated and is assigned to a variable hra
    ta=(5/100)*mbs # 5% of mbs is calculated and is assigned to a variable ta
    gs=(mbs+hra+ta)*12 # mbs, hra and ta are added and the result is multiplied by 12 and is 
                       # assigned to a variable 'gs'
    
    print('\nHRA :',hra)
    print('TA :',ta)
    print("SALARY STATUS : Less than 4000") # Prints the salary status
    print("\nANNUAL GROSS SALARY :",'$',round(gs,2)) # Prints the value of gs
    
elif mbs>4000 and mbs<=10000: # if : Checks if the value of mbs satisfies the conditions 
                              # greater than $4000 and lesser than or equal to $10000 
                              # using AND operator
            
    hra=(10/100)*mbs # 10% of mbs is calculated and is assigned to a variable hra
    ta=(10/100)*mbs # 10% of mbs is calculated and is assigned to a variable ta
    gs=(mbs+hra+ta)*12 # mbs, hra and ta are added and the result is multiplied by 12 and is 
                       # assigned to a variable 'gs'
    
    print('\nHRA :',hra)
    print('TA :',ta)
    print("SALARY STATUS : Greater than 4000 and less than or equal to 10000") 
                                                        # Prints the salary status
    print("\nANNUAL GROSS SALARY :",'$',round(gs,2)) # Prints the value of gs
    
elif mbs>10000: # if : Checks if the value of mbs satisfies the condition 
                # greater than $10000 using AND operator
        
    hra=(15/100)*mbs # 15% of mbs is calculated and is assigned to a variable hra
    ta=(15/100)*mbs # 15% of mbs is calculated and is assigned to a variable ta
    gs=(mbs+hra+ta)*12 # mbs, hra and ta are added and the result is multiplied by 12 and is 
                       # assigned to a variable 'gs'
    print('\nHRA :',hra)
    print('TA :',ta)
    print("SALARY STATUS : Greater than 10000") # Prints the salary status
    print("\nANNUAL GROSS SALARY :",'$',round(gs,2))# Prints the value of gs
    
else: # else : Executes if the if condition is False
    print('INVALID SALARY. RETRY.') # Prints if the else condition is True
    

'''
We are accepting the basic monthly salary from the user and calculating the 
hra, ta and gross salary according to the given conditions.
'''
```

```python
# TASK5

a = input("Enter an Alphabet : ") # Accepts a string character and stores it in the variable 'a'

if a.isalpha(): # Checks if the entered character is a alphabet
    
    if a.isupper(): # Checks if the alphabet is in uppercase
        print("UPPERCASE") # Prints if the if condition is True
        
    else: # else : Executes if the if condition is False
        print("lowercase") # Prints if the else condition is True
        
else: # else : Executes if the if condition is False
    print("INVALID. RETRY.") # Prints if the else condition is True
    
'''
We are accepting an alphabet and checking if it is an alphabet and in uppercase 
using isalpha() and isupper(). If True the output will be 'UPPERCASE', if False the output
will be 'lowercase'.
If it is not an alphabet the output will be 'INVALID. RETRY.'
'''
```
