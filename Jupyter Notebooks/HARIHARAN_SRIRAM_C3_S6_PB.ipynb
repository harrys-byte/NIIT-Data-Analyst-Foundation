```python
print("HARRY POTTER QUIZ TRIVIA".center(94))
print("\nTOTAL NUMBER OF QUESTIONS : 10")
print("TOTAL MARKS : 20 MARKS")
print("CORRECT ANSWER : 2 MARKS")
print("WRONG ANSWER : NO MARKS")
print("ENTER 'S' IN ORDER TO SKIP A QUESTION | NO MARKS WILL BE AWARDED\n")

mcq={'Q1 : What spell did Harry use to kill Lord Voldemort?': ['A. Expelliarmus','B. Expecto Patronum','C. Avada Kedavra','D. Accio'],
    'Q2 : What is a thestral?': ['A. A half-giant','B. An invisible winged horse',
                                 'C. A shrunken head','D. A pixie'],
    'Q3 : What was the name of the joke shop founded by the Weasley twins at 93 Diagon Alley? : ':
    ['A. Weasley’s Witchcraft Wonders','B. Weasley’s Worldwide Whompers',
     'C. Weasley’s Wicked Whatsits','D. Weasley’s Wizard Wheezes'],
    'Q4 : Finish the inscription on Dobby’s tombstone: ‘Here lies Dobby…':['A. A true friend',
'B. The best servant','C. A free Elf','D. Master of socks'],
    'Q5 : What magical talent does Harry share with Voldemort?':['A. Animagus','B. Parselmouth',
'C. Being an Auror','D. Being a Death Eater'],
    'Q6 : What position does Harry play on his Quidditch team?':['A. Chaser','B. Keeper',
'C. Bludger','D. Seeker'],
    'Q7 : The tears of which animal are the only known antidote to basilisk venom?':['A. Phoenix',
'B. Billywig','C. Hippogriff','D. Boggart'],
    'Q8 : When unearthed, a mandrake will do what?' : ['A. Dance','B. Burp','C. Scream','D. Laugh'],
    'Q9 : Cedric Diggory faced what breed of dragon in the Triwizard Tournament?' : [
     'A. Swedish Short-Snout','B. Peruvian Vipertooth','C. Common Welsh Green',
        'D. Norwegian Ridgeback'],
    'Q10 : ‘Wit beyond measure is man’s greatest treasure’ is the motto of which house?':[
        'A. Gryffindor','B. Hufflepuff','C. Ravenclaw','D. Slytherin']}

crctop=['A','B','D','C','B','D','A','C','A','C']
score=0
i=0
sk=cr=wr=0

for k,v in mcq.items():
    print(k)
    for j in v:
        print(j)
    
    a=input("Enter Option : ")
    
    if a.upper()=='S':
        i+=1
        sk+=1
        print('\nQuestion Skipped\n')
        continue;
        
    if a.upper()==crctop[i]:
        print("\nCorrect Answer\n")
        score+=2
        cr+=1
    else:
        print("\nWrong Answer\n")
        wr+=1
    i+=1

print('\n')
print("OVERALL ANALYSIS".center(94))
print('\nFINAL CUMMULATIVE SCORE :',score)
print("FINAL PERCENTAGE :",round((score/20)*100,2),'%')
print('\nNUMBER OF QUESTIONS ANSWERED CORRECT:',cr)
print('NUMBER OF QUESTIONS ANSWERED WRONG:',wr)
print('NUMBER OF QUESTIONS SKIPPED :',sk)

```
